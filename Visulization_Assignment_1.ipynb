# Flight

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# IMPORTING DATASETS
# Flight Dataset
flight_df = pd.read_excel('flight_price.xlsx')
# Playstore Dataset
play_store_df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
```

```python
# Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?

# GETTING DIMENTIONS OF THE DATASET
num_rows = flight_df.shape[0]
num_columns = flight_df.shape[1]

print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
# Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

# PLOTTING HISTOGRAM TO UNDERSTAND DISTRIBUTION OF FLIGHT PRISES
plt.figure(figsize=(10, 6))
sns.histplot(flight_df['Price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

# Calculate the range of prices
price_range = flight_df['Price'].max() - flight_df['Price'].min()

# Get the minimum and maximum prices
min_price = flight_df['Price'].min()
max_price = flight_df['Price'].max()

print("Range of prices:", price_range)
print("Minimum price:", min_price)
print("Maximum price:", max_price)
```

```python
# Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

# Create a boxplot of flight prices by airline
plt.figure(figsize=(12, 8))
sns.boxplot(x='Airline', y='Price', data=flight_df)
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

# Create a boxplot of flight prices
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price', data=flight_df)
plt.title('Flight Prices - Boxplot')
plt.xlabel('Price')
plt.show()

# Outliers, deviating significantly from the dataset's main distribution, can distort summary statistics and relationships, leading to biased models and misinterpretations.
# They skew measures of central tendency and spread, potentially misleading analysis conclusions. Ignoring or mishandling outliers can compromise the reliability and robustness
# of statistical analyses and machine learning models.
```

```python
# Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season.
# What features would you analyze to identify the peak season, and how would you present your findings to your boss?

# To identify the peak travel season from the Flight Price dataset, I would analyze several key features:
#  1. Date_of_Journey: This column provides the date of the flight journey. By aggregating flight prices over different months or quarters,
#     we can identify periods of higher demand and potentially peak travel seasons.

# Creating new copied data frame for our operations and maintaining our orginality of of initial flight dataset
copy_flight_df = flight_df.copy()

# Convert Date_of_Journey to datetime
copy_flight_df['Date_of_Journey'] = pd.to_datetime(copy_flight_df['Date_of_Journey'])

# Extract month from Date_of_Journey
copy_flight_df['Journey_Month'] = copy_flight_df['Date_of_Journey'].dt.month

# Plot average flight prices by month
plt.figure(figsize=(10, 6))
sns.lineplot(x='Journey_Month', y='Price', data=copy_flight_df)
plt.title('Average Flight Prices by Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(range(1, 13))
plt.show()
```

```python
# Q6 Continued...
# 2. Destination: Considering the destinations of the flights and their respective peak tourist seasons can provide insights into when travelers are
# likely to fly to specific locations.

plt.figure(figsize=(12, 8))
sns.boxplot(x='Destination', y='Price', data=flight_df)
plt.title('Flight Prices by Destination')
plt.xlabel('Destination')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# Q6 Continued...
# 3. Total_Stops: The number of stops in a flight route can also impact travel demand and prices.
#    Analyzing flight prices based on the number of stops can help identify preferences for direct flights or layovers during peak travel seasons.

plt.figure(figsize=(8, 6))
sns.barplot(x='Total_Stops', y='Price', data=flight_df)
plt.title('Flight Prices by Total Stops')
plt.xlabel('Total Stops')
plt.ylabel('Average Price')
plt.show()
```

```python
# Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories.
# How would you approach this task, and what features would you analyze to make recommendations to the company?


# 1. Analyze install counts by category
# Shows the total number of installs for apps in each category.
# Helps identify categories with the highest user engagement and popularity.

plt.figure(figsize=(20, 6))
sns.barplot(x='Category', y='Installs', data=play_store_df, estimator=sum, ci=None, order=play_store_df.groupby('Category')['Installs'].sum().sort_values(ascending=False).index)
plt.title('Total Installs by Category')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app.
# What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

# Adding 2 new col : 'Year', 'Month'
# play_store_df['Last Updated'].unique()
# Define a function to extract the year
def extract_year(date):
    split_date = date.split(', ')
    if len(split_date) > 1:
        return split_date[1]
    else:
        return ''

# Define a function to extract the month
def extract_month(date):
    split_date = date.split(' ')
    if len(split_date) > 0:
        return split_date[0]
    else:
        return ''

# Apply the functions to create 'Year' and 'Month' columns
play_store_df['Year'] = play_store_df['Last Updated'].apply(extract_year)
play_store_df['Month'] = play_store_df['Last Updated'].apply(extract_month)

play_store_df
```

```python
# month of update
update_month = pd.DataFrame(play_store_df['Month'].value_counts()).reset_index()
update_month.columns = ['Month', 'Updates']
update_month
```

```python
# Define the order for the months
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 4))  # Use only one subplot

# Bar plot
ax = sns.barplot(data=update_month, x='Updates', y='Month', order=months)
ax.set_title('Month wise no. of updates')

# Add labels to the bars
for i in ax.containers:
    ax.bar_label(i)

plt.show()
```
